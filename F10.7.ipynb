# Daily flux values
## Daily flux values for the year 2021

website: https://spaceweather.gc.ca/forecast-prevision/solar-solaire/solarflux/sx-5-flux-en.php

```python
url = 'https://spaceweather.gc.ca/forecast-prevision/solar-solaire/solarflux/sx-5-flux-en.php'

r = requests.get(url)
s = BeautifulSoup(r.text, "html.parser")
#data = s.find_all("tbody")

table = s.find('table')


table
```

```python
table_body = table.find_all('tr')

cols_name = table_body[0].text.split('\n')

data = []

for num in range(1, len(table_body)):
    result = table_body[num].text.split('\n')
    data.append(result)
    
```

```python
df = pd.DataFrame(data, columns = cols_name)
df['datetime'] = df['Date'] +' '+ df['Time']
df.index = pd.to_datetime(df['datetime'])

df.drop(["Date", 'Time', 'Carringtonrotation', 
         'datetime', 'Julian day', ''], axis=1, inplace=True)

for col in list(df.columns):
    df[col] = pd.to_numeric(df[col])
    
    
df.to_csv("DialyFlux.txt", sep = ' ')
```
